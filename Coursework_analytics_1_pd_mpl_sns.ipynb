## 1. Описание и начальная работа с данными

### 0. Подготовка данных

```python
# импортируем необходимые библиотеки

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# считаем данные из файлов 

courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv', parse_dates=['birthday'], dayfirst=False)
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv', parse_dates=['start_date', 'finish_date'], dayfirst=False)
```

```python
# объединим датасеты в 1

df = progress_phases.merge(progresses, how='left', left_on='progress_id', right_on='id')
df = df.drop(columns=['id'])
df = df.merge(courses, how='left', left_on='course_id', right_on='id')
df.drop(columns=['Unnamed: 0', 'id'], inplace=True)
df = df.merge(course_contents, how='left', on=['course_id', 'module_number', 'lesson_number'])
df
```

## 2. Расчет потенциальной нагрузки на преподавателей

### 2.1. Количество студентов в каждом месяце для каждого курса, которые начинали выполнение первой домашней работы в каждом месяце.

```python
# оставим только те модули и уроки, где есть дз, отсортируем по номеру модуля и урока и возьмем самое первое значение для каждого курса
first_hw = df[df['is_homework'] == True
             ].sort_values(by=['module_number', 'lesson_number']
                          ).groupby('title')[['module_number', 'lesson_number']
                                            ].agg(lambda x: x.iloc[0]).reset_index()

# переименуем колонки
first_hw.columns = ['title', 'first_hw_mod', 'first_hw_les']

#  добавим полученную колонку с первым модулем с дз в каждом курсе в изначальный датафрейм
df = df.merge(first_hw, how='left', on='title')
```

### 2.4. Количество прогрессов по выполнению домашних работ в каждом месяце.

```python
# определим самую позднюю дату в датасете, извлечем из нее год и месяц и приведем ее к формату Timestamp
last_date = max(df['start_date'].max(), df['finish_date'].max())
last_date = pd.to_datetime('-'.join([str(last_date.year), str(last_date.month)]))
last_date
```

```python
# на основе колонок начала и окончания выполнения урока создадим два столбца, содержащих только год и месяц

df['start_year_month'] = pd.to_datetime(df['start_date'].dt.strftime('%Y-%m'))
df['finish_year_month'] = pd.to_datetime(df['finish_date'].dt.strftime('%Y-%m'))
```

```python
# с помощью функции data_range, которая заменяет отсутствующие даты окончания на самую позднюю дату last_date, 
#     определяет частоту freq='M' и создает лист DatetimeIndex, создадим новый столбец, 
#     содержащий последовательность месяцев от начала до окончания выполнения урока

def date_range(start, end):
    if end != end:
        end = last_date
    return pd.date_range(start, end + pd.tseries.offsets.MonthEnd(), freq='M')

hw_to_check = df.copy()
hw_to_check['date_range'] = list(map(date_range, hw_to_check['start_year_month'], hw_to_check['finish_year_month']))
```

```python
# разделим строки на несколько строк по спискам столбца date_range

hw_to_check = hw_to_check.explode('date_range')
```

```python
# отфильтруем уроки, где есть домашнее задание, сгруппируем прогрессы по курсам и месяцам в столбце date_range и посчитаем их количество

hw_to_check = hw_to_check[hw_to_check['is_homework'] == True].groupby(['title', 'date_range'])['progress_id'].count().reset_index()
hw_to_check.columns = ['course', 'month', 'homework_quantity']
hw_to_check.info()
```

## 3. Выявление проблемных модулей

### 3.1. Время прохождения каждого модуля 

```python
# очистим датасет от записей, не имеющих даты окончания выполнения дз, а также отфильтруем только уроки и модули, где есть дз 
df_cleaned = df[(df['finish_date'] == df['finish_date']) & (df['is_homework'] == True)]
```

```python
# создадим столбец с разностью времени начала и окончания выполнения дз
df_cleaned['diff'] = df['finish_date'] - df['start_date']
```

```python
# отфильструем данные, где дата начала модуля была позднее даты окончания
df_cleaned = df_cleaned[df_cleaned['diff'] >= pd.Timedelta(0)]
```

```python
# сгруппируем по курсам, модулям и студентом, суммируя разности столбцов окончания и начала выполнения дз (таким образом, если в модуле несколько дз, мы получим суммарное время прохождения 1 модуля одним человеком)
df_grouped = df_cleaned.groupby(['title', 'module_number', 'student_id'])['diff'].sum().reset_index()

# сгруппируем по курсам и модулям и опишем полученные датасет с помощью метода describe
df_grouped = df_grouped.groupby(['title', 'module_number']).describe().reset_index()

# переименуем столбцы, убирая мультииндексацию
df_grouped.columns = ['course', 'module', 'count', 'mean', 'std', 'min', '25%', 'median', '75%', 'max']
df_grouped
```

### 3.2. Медианное время прохождения каждого модуля для каждого курса (15 графиков)

```python
import numpy as np

# подготовим данные для графика: извлечем из датасета только столбец, соответствующий медиане, переведем время в дни
median_time = df_grouped.loc[:, ['course', 'module', 'median']]
median_time['median'] = median_time['median'].apply(lambda x: x.total_seconds() / (60 * 60 * 24))
median_time.info()
```

```python
# построим 15 графиков с помощью seaborn.FacetGrid и plt.plot

g = sns.FacetGrid(median_time, col='course', col_wrap=3, aspect=1.5)
g.map(sns.lineplot, 'module', 'median')

for i, ax in enumerate(g.axes):
    if i in (12, 13, 14):
        ax.set_xticks(list(range(0, 22, 2)))

plt.suptitle('Медианное время прохождения каждого модуля курса', y=1.03)
plt.show()
```

### 3.3. Медианное время выполнения домашней работы по месяцам.

```python
# выделим в отдельный столбец месяц из даты старта домашней работы в отфильтрованном датасете df_cleaned
#   (очищенном от записей, не имеющих даты окончания прохождения урока, не имеющих дз, а также от данных,
#   где дата начала модуля была позднее даты окончания)

df_cleaned['month'] = df_cleaned['start_date'].dt.month
```

## 4. Расчет конверсии

### 4.1. Конверсия перехода студентов из одного модуля в другой

```python
# отфильтрум датасет по модулям и урокам, имеющим дз, оставим только курс, модуль, урок, студента, и статус и упорядочим по курсу, модулю, уроку
df_filtered = df[df['is_homework'] == True].loc[:, ['title', 'module_number', 'lesson_number', 'student_id', 'status']].sort_values(by=['title', 'module_number', 'lesson_number'])
df_filtered
```

```python
# посчитаем для каждого курса и модуля, сколько человек его закончило, отфильтруем по done (для определения последнего дз предварительно сгруппируем датасет по курсу, модулю, студенту)
df_finished = df_filtered.groupby(['title', 'module_number', 'student_id'])[['lesson_number', 'status']].first().reset_index()
df_finished = df_finished[df_finished['status'] == 'done'].groupby(['title', 'module_number'])['status'].count().reset_index()
df_finished[df_finished['title'] == 'PHP-разработчик с 0 до PRO. Часть 1']
```

```python
# посчитаем для каждого курса и модуля, сколько человек его начало (для определения первого дз предварительно сгруппируем датасет по курсу, модулю, студенту)
df_started = df_filtered.groupby(['title', 'module_number', 'student_id'])[['lesson_number', 'status']].last().reset_index()
df_started = df_started.groupby(['title', 'module_number'])['status'].count().reset_index()
df_started[df_started['title'] == 'PHP-разработчик с 0 до PRO. Часть 1']
```

```python
# добавим в датасет колонку с кол-вом по предыдущему модулю - merge_asof(allow_exact_matches=False)
df_conv = pd.merge_asof(df_started.sort_values(by='module_number'), df_finished.sort_values(by='module_number'), direction='backward', on='module_number', by='title', allow_exact_matches=False)
df_conv.columns = ['course', 'module', 'people_started_this', 'people_finished_prev']
```

```python
# добавим столбец с рассчитанной конверсией
df_conv['conversion'] = df_conv['people_started_this'] / df_conv['people_finished_prev']
df_conv
```

## 5. Метрика успеваемости 

&nbsp;&nbsp;&nbsp;&nbsp;

Иногда студенты берут курсы, которые оказываются для них неподъёмными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумаем и обоснуем метрику успеваемости студента.

Возможные метрики:
1. Время прохождения модуля студентом в сравнении с 90-процентилем времени прохождения модуля. - при условии, что все модули в статусе 'done', скорее скажет о том, что курс дается тяжело (обучение идет медленно), а не о низкой мотивации.
2. Процент начатых, но не завершенных успешно модулей (start/fail), от всех начатых модулей - может говорить также о том, что курс оказался слишком тяжелым для студента - есть попытки учиться, но они не заканчиваются успешно
3. Успешно пройдено 25% курса в адекватное время (до 75-процентиля), а последний модуль или в статусе start уже долгое время (дольше 75-процентиля) или вообще не начат - говорит о том, что курс был посильным, однако мотивация стала недостаточной.

Наиболее простой и четко описывающей необходимый результат представляется метрика номер 2, для 1 и 3 нужно более детально продумывать отсечки по времени прохождения и количеству пройденных модулей.

Рассчитаем процент начатых, но не завершенных успешно модулей для каждого студента. Однако отбросим последние для каждого студента и курса работы в статусе "start", если они были начаты позднее 1 мая 2019 года (датасет заканчивается июлем 2019 года) - они могут быть просто в процессе выполнения. (Возьмем только последние работы каждого студента, чтобы не отбрасывать возможных проблемных студентов последних месяцев)

```python
# добавим столбец для фильтрации 
df_filter = df[df['status'] == 'start'].sort_values(by='start_date', ascending=True).groupby(['student_id', 'title'])[['start_date', 'progress_id', 'module_number']].last().reset_index(allow_duplicates=True)
df_filter['not_count'] = np.where(df_filter['start_date'].dt.tz_convert(None) > pd.to_datetime('2019-05-01'), 1, 0)

# добавим столбец not_count в общий датафрейм, чтобы обозначить записи, 
#         которые мы не хотим учитывать при расчете - работы, начатые (в статусе start) позднее 1 мая 2019.

df = df.merge(df_filter[['student_id', 'title', 'start_date', 'not_count']], how='left', on=['student_id', 'title', 'start_date']).fillna(0)
df
```

```python
# рассчитаем все не законченные работы по студентам и курсам (статусы start и fail)
df_bad = df[(df['status'].isin(['start', 'fail'])) & (df['not_count'] == 0)].groupby(['student_id', 'title'])['status'].count().reset_index()
df_bad.columns = ['student', 'course', 'not_finished_lessons']
display(df_bad)

# рассчитаем кол-во всех работ для каждого студента по курсам
df_all = df[(df['not_count'] == 0)].groupby(['student_id', 'title'])['status'].count().reset_index()
df_all.columns = ['student', 'course', 'started_lessons']
display(df_all)
```

```python
# объединим датасеты, заполняя NaN в колонке с не законченными модулями нулями
df_all = df_all.merge(df_bad, how='left', on=('course', 'student')).fillna(0)
```

```python
# рассчитаем количество уроков в каждом курсе
df_lesson_count = df.groupby(['title', 'module_number'])['lesson_number'].nunique().reset_index().groupby('title')['lesson_number'].sum().reset_index()
df_lesson_count.columns = ['course', 'total_lessons']

# добавим столбец с общим кол-вом уроков по курсу в датафрейм с подсчетом законченных и не законченных работ студентами
df_all = df_all.merge(df_lesson_count, how='left', on='course')
df_all
```

```python
df_all.info()
```

```python
# рассчитаем процент начатых, но не законченных уроков от всех начатых уроков- чем ниже процент, тем вероятнее, что курс дается студенту хорошо
df_all['percent_not_finished_from_all_started'] = (df_all['not_finished_lessons'] / df_all['started_lessons'] * 100).round(2)

# рассчитаем процент уроков, которые студент вообще начинал (не имеет значения статус), от всех уроков курса - чем выше процент, тем вероятнее, что курс дается студенту хорошо
df_all['percent_started_from_total_lessons'] = (df_all['started_lessons'] / df_all['total_lessons'] * 100).round(2)
df_all
```

Основной расчет был сделал, теперь важно верно выбрать отсечки, чтобы отразить именно тех студентов, которых мы считаем проблемными. Предлагается выбрать студентов, которые не завершили 50% работ, но при этом начали более 2. Таким образом отсеем тех людей, которые начали пока слишком мало работ, чтобы судить о том, что им не удается их закончить.

```python
# отберем тех у кого процент не законченных от всех начатых работ выше 50% и количество начатых уроков больше 2
troubles = df_all[(df_all['percent_not_finished_from_all_started'] >= 50) & (df_all['started_lessons'] > 2)]

# отсортируем по частоте встречаемости в датасете проблемных студентов
sorted_courses = troubles['course'].value_counts().index.tolist()
troubles = troubles.sort_values(by='course', key=lambda x: x.map({val: i for i, val in enumerate(sorted_courses)}))

# добавим сферы курсов
troubles = troubles.merge(df[['title', 'field']].drop_duplicates(), how='left', left_on='course', right_on='title')
```

```python
sns.histplot(troubles, y='course', hue='field', palette='magma', alpha=0.7)
plt.title('Количество проблемных студентов в зависимости от курса', loc='left', weight='bold')
plt.xlabel('Кол-во студентов', loc='left')
plt.ylabel('')

plt.show()
```
